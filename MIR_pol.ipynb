# Mid-infrared polarimetry: a Bayesian approach 

The aim of this code is to develop a Bayesian approach to infere the absorptive and emissive components of a given observations. In this case the observatiosn are simulated with the already known A,B,C,D parameters

```python
%matplotlib inline
import os
from pymc3 import  *
from pymc3.stats import *
import numpy as np
import matplotlib.pyplot as plt
from readcol import readcol
import patsy
import theano
from pandas import *
from statsmodels.formula.api import glm as glm_sm
import statsmodels.api as s
import numpy as np
from scipy import stats
```

##### Reading observational data

```python
#BN absorptive component use for modeling from Aitken+04 (MNRAS, 348, 279)
l_bn,fa_bn=readcol("BN_fa.dat",twod=False)
l_bn,fe_bn=readcol("BN_fe.dat",twod=False)

#Assumed extintion Rv=5.5 from Draine:  
#ftp://ftp.astro.princeton.edu/draine/dust/mix/kext_albedo_WD_MW_5.5A_30_D03.all
l_ext,tau_ext=readcol("ext_Rv55.dat",twod=False)

#BN emissive component use for modeling
#fe_bn = fa_bn/tau_ext
#fe_bn = fe_bn/max(fe_bn)

#Imaging BN data from Aitken+2004 (MNRAS, 348, 279)
l_bn_ima,Dx_ima,fa_ima,fe_ima=readcol("BN.dat",twod=False)
Dx_ima=0. #I do not want to show the x-uncertainty yet.

###############
### Figures ###
###############
font_size=6.0
label_size=6.
lw = 3.
fig,(ax1,ax2) = plt.subplots(1,2,figsize=(18.0,6.),dpi=120)

#Imaging data
ax1.errorbar(l_bn_ima,fa_ima,xerr=Dx_ima,fmt='o',label='Absorptive component',color='blue')
ax1.errorbar(l_bn_ima,fe_ima,xerr=Dx_ima,fmt='o',label='Emissive component',color='red')

ax1.set_xlabel('Wavelength ($\mu$m)')
ax1.set_ylabel('Polarization normalized to unity')
ax1.set_ylim([0,1])
ax1.set_xlim([7,14])
ax1.legend(loc='best')

#Model
ax2.plot(l_bn,fa_bn,color='black',lw=lw,label='Model: absorptive component')
ax2.plot(l_bn,fe_bn,color='black',lw=lw,linestyle='-.',label='Model: emissive component')

ax2.set_xlabel('Wavelength ($\mu$m)')
ax2.set_ylabel('Polarization normalized to unity')
ax2.set_ylim([0,1])
ax2.set_xlim([7,13])
ax2.legend(loc='best')
```

```python
#Simulate observations
#In this case the linear combination has some known factors (A,B,C,D) to test the Bayesian approach. 
#q_obs and u_obs: to be changed by the real observations.
A_obs = 0.02
B_obs = 0.001
C_obs = 0.001
D_obs = 0.01

qerr = np.random.rand(len(l_bn_ima))/100.
uerr = np.random.rand(len(l_bn_ima))/100.

qa_obs = A_obs*fa_ima
qe_obs = B_obs*fe_ima
q_obs  = qa_obs + qe_obs

ua_obs = C_obs*fa_ima
ue_obs = D_obs*fe_ima
u_obs  = ua_obs + ue_obs

p_a_obs = 100.*np.sqrt(qa_obs*qa_obs + ua_obs*ua_obs)
p_e_obs = 100.*np.sqrt(qe_obs*qe_obs + ue_obs*ue_obs)
p_obs   = 100.*np.sqrt(q_obs*q_obs + u_obs*u_obs)

pa_a_obs = 0.5*np.arctan(ua_obs/qa_obs)*180./3.1415 
pa_e_obs = 0.5*np.arctan(ue_obs/qe_obs)*180./3.1415 
pa_obs   = 0.5*np.arctan(u_obs/q_obs)*180./3.1415 

###############
### Figures ###
###############
font_size=6.0
label_size=6.
lw = 3.
fig,(ax1,ax2,ax3,ax4) = plt.subplots(1,4,figsize=(25.0,5.),dpi=120)

#Imaging data
ax1.errorbar(l_bn_ima,q_obs,xerr=Dx_ima,yerr=qerr,fmt='o',label='q',color='black')
ax1.plot(l_bn_ima,qa_obs,linestyle='-',color='blue')
ax1.plot(l_bn_ima,qe_obs,linestyle=':',color='blue')

#q
ax1.set_xlabel('Wavelength ($\mu$m)')
ax1.set_ylabel('q')
ax1.set_ylim([-0.05,0.05])
ax1.set_xlim([7,14])
ax1.legend(loc='best')

#u
ax2.errorbar(l_bn_ima,u_obs,xerr=Dx_ima,yerr=uerr,fmt='o',label='u',color='black')
ax2.plot(l_bn_ima,ua_obs,linestyle='-',color='blue')
ax2.plot(l_bn_ima,ue_obs,linestyle=':',color='blue')

ax2.set_xlabel('Wavelength ($\mu$m)')
ax2.set_ylabel('u')
ax2.set_ylim([-0.05,0.05])
ax2.set_xlim([7,14])
ax2.legend(loc='best')

#p
ax3.errorbar(l_bn_ima,p_obs,xerr=Dx_ima,fmt='o',label='P',color='black')
ax3.plot(l_bn_ima,p_a_obs,linestyle='-',color='blue')
ax3.plot(l_bn_ima,p_e_obs,linestyle=':',color='blue')

ax3.set_xlabel('Wavelength ($\mu$m)')
ax3.set_ylabel('Degree of polarization')
ax3.set_ylim([0,10])
ax3.set_xlim([7,14])
ax3.legend(loc='best')

#pa
ax4.errorbar(l_bn_ima,pa_obs,xerr=Dx_ima,fmt='o',label='PA',color='black')
ax4.plot(l_bn_ima,pa_a_obs,linestyle='-',color='blue')
ax4.plot(l_bn_ima,pa_e_obs,linestyle=':',color='blue')

ax4.set_xlabel('Wavelength ($\mu$m)')
ax4.set_ylabel('P.A. of polarization')
ax4.set_ylim([-90,90])
ax4.set_xlim([7,14])
ax4.legend(loc='best')
```

##### Inference Model

```python
#The aim is to fit q and u (as shown in the above plots) using the following linear combination 
#of 4 independent parameters (A,B,C,D) and 2 already known empirical functions (fa, fe).
#q = A*fa*lambda + B*fe*lambda
#u = C*fa*lambda + D*fe*lambda

#Boundaries of A,B,C,D factors by n_minmax time to the min. and max. of q and u.
n_minmax = 5
qmin = min(q_obs)
umin = min(u_obs)
if umin < 0: 
    u_lower = umin + n_minmax*umin
else:
    u_lower = qmin - n_minmax*umin
if qmin < 0:
    q_lower = umin + n_minmax*qmin
else:
    q_lower = qmin - n_minmax*qmin
    
q_upper = max(q_obs) + n_minmax*max(q_obs)
u_upper = max(u_obs) + n_minmax*max(u_obs)

#Built the model
with Model() as hierarchical_model:
    #Variables are Uniform functions with width equal to 5 x std(q,u)
    A  = Uniform('A',lower=q_lower,upper=q_upper)
    B  = Uniform('B',lower=q_lower,upper=q_upper)
    C  = Uniform('C',lower=u_lower,upper=u_upper)
    D  = Uniform('D',lower=u_lower,upper=u_upper)
    #Deterministic model. Dependence of variables
    qa = A*fa_ima*l_bn_ima
    qe = B*fe_ima*l_bn_ima
    ua = C*fa_ima*l_bn_ima
    ue = D*fe_ima*l_bn_ima
    q_fit = qa + qe
    u_fit = ua + ue
    # Data likelihood. Normal distributions per each observable (data point per wavelength) 
    #with stdv the error of each data point from the observations.
    q_like  = Normal('q', mu=q_fit, sd=qerr, observed=q_obs) 
    u_like  = Normal('u', mu=u_fit, sd=uerr, observed=u_obs)

with hierarchical_model:    
    #Inference
    start = find_MAP()
    step = NUTS([A,B,C,D,q_like,u_like],scaling=start)
    # draw n_post posterior samples using NUTS sampling
    n_post = 25000
    hierarchical_trace = sample(n_post, step, start=start) 
```

```python
#############
### PLOTS ###
#############
#Remove the first n_steps from the trace
n_steps = 5000
#Plots of the posteriors and trace to check convergence beyond n_step.
traceplot(hierarchical_trace[n_steps:],['A','B','C','D']);    
```

```python
summary(hierarchical_trace[n_steps:],['A','B','C','D'])
```

##### Posteriors and fitting

```python
#Data from trace to dataframe using pandas
df_trace = trace_to_dataframe(hierarchical_trace[n_steps:])

################
### Best fit ###
################
A_m     = stats.mode(df_trace['A'])[0]
B_m     = stats.mode(df_trace['B'])[0]
C_m     = stats.mode(df_trace['C'])[0]
D_m     = stats.mode(df_trace['D'])[0]

fa = fa_ima*l_bn_ima
fe = fe_ima*l_bn_ima

qa_bestfit = A_m*fa
qe_bestfit = B_m*fe
q_bestfit = qa_bestfit + qe_bestfit

ua_bestfit = C_m*fa
ue_bestfit = D_m*fe
u_bestfit = ua_bestfit + ue_bestfit


p_bestfit  = 100.*np.sqrt(q_bestfit*q_bestfit + u_bestfit*u_bestfit)
p_a_bestfit = 100.*np.sqrt(qa_bestfit*qa_bestfit + ua_bestfit*ua_bestfit) 
p_e_bestfit = 100.*np.sqrt(qe_bestfit*qe_bestfit + ue_bestfit*ue_bestfit) 

pa_bestfit   = 0.5*np.arctan(u_bestfit/q_bestfit)*180./3.1415 
pa_a_bestfit = 0.5*np.arctan(ua_bestfit/qa_bestfit)*180./3.1415 + zeros(len(l_bn_ima))
pa_e_bestfit = 0.5*np.arctan(ue_bestfit/qe_bestfit)*180./3.1415 + zeros(len(l_bn_ima))


##################
### Posteriors ###
##################
A_max = hpd(hierarchical_trace['A'])[1]
A_min = hpd(hierarchical_trace['A'])[0]
B_max = hpd(hierarchical_trace['B'])[1]
B_min = hpd(hierarchical_trace['B'])[0]
C_max = hpd(hierarchical_trace['C'])[1]
C_min = hpd(hierarchical_trace['C'])[0]
D_max = hpd(hierarchical_trace['D'])[1]
D_min = hpd(hierarchical_trace['D'])[0]


###############
### Figures ###
###############
font_size=6.0
label_size=6.
lw = 3.
ll = l_bn_ima
fig,(ax1,ax2,ax3,ax4) = plt.subplots(1,4,figsize=(25.0,5.),dpi=120)

for ii in range(0,1000):
    qa = df_trace.A[ii] * fa 
    qe = df_trace.B[ii] * fe
    q = qa + qe
    ax1.plot(ll,q,lw=1,color='red',alpha=.01)

    ua = df_trace.C[ii] * fa 
    ue = df_trace.D[ii] * fe
    u = ua + ue
    ax2.plot(ll,u,lw=1,color='red',alpha=.01)
    
    p = 100*np.sqrt(q*q + u*u)
    ax3.plot(ll,p,lw=1,color='red',alpha=.01)
    
    pa = 0.5*np.arctan(u/q)*180./3.1592
    ax4.plot(ll,pa,lw=1,color='red',alpha=.01)

#q
ax1.errorbar(l_bn_ima,q_obs,yerr=qerr,fmt='o',label='q',color='blue')
ax1.plot(ll,q_bestfit,lw=2,color='black',label='Model: total')
ax1.plot(ll,qa_bestfit,lw=2,color='black',linestyle='-.',label='Model: absorptive component')
ax1.plot(ll,qe_bestfit,lw=2,color='black',linestyle=':',label='Model: emissive component')

ax1.set_xlabel('Wavelength ($\mu$m)')
ax1.set_ylabel('q')
ax1.set_ylim([-0.05,0.05])
ax1.set_xlim([7,14])


#u
ax2.errorbar(l_bn_ima,u_obs,yerr=uerr,fmt='o',label='u',color='blue')
ax2.plot(ll,u_bestfit,lw=2,color='black',label='Model: total')
ax2.plot(ll,ua_bestfit,lw=2,color='black',linestyle='-.',label='Model: absorptive component')
ax2.plot(ll,ue_bestfit,lw=2,color='black',linestyle=':',label='Model: emissive component')


ax2.set_xlabel('Wavelength ($\mu$m)')
ax2.set_ylabel('u')
ax2.set_ylim([-0.05,0.05])
ax2.set_xlim([7,14])

#p
ax3.errorbar(l_bn_ima,p_obs,fmt='o',label='P',color='blue')
ax3.plot(ll,p_bestfit,lw=2,color='black',label='Model: total')
ax3.plot(ll,p_a_bestfit,lw=2,color='black',linestyle='-.',label='Model: absorptive component')
ax3.plot(ll,p_e_bestfit,lw=2,color='black',linestyle=':',label='Model: emissive component')


ax3.set_xlabel('Wavelength ($\mu$m)')
ax3.set_ylabel('Degree of polarization (%)')
ax3.set_ylim([0,10])
ax3.set_xlim([7,14])

#pa
ax4.errorbar(l_bn_ima,pa_obs,fmt='o',label='PA',color='blue')
ax4.plot(ll,pa_bestfit,lw=2,color='black',label='Total Model')
ax4.plot(ll,pa_a_bestfit,lw=2,color='black',linestyle='-.',label='absorptive component')
ax4.plot(ll,pa_e_bestfit,lw=2,color='black',linestyle=':',label='emissive component')

ax4.set_xlabel('Wavelength ($\mu$m)')
ax4.set_ylabel('P.A. of polarization')
ax4.set_ylim([-90,90])
ax4.set_xlim([7,14])
ax4.legend(loc='upper right')


## Save Fig. to PDF and display
plt.savefig('MIR_model.pdf')
#os.system('open MIR_model.pdf')
```
